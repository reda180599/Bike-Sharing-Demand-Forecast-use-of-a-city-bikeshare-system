### Create or modify new features

```python
# Log normalize the feature 'Count' to convert it to normal distribution

df1 = data_copy['count']

df2 = np.log(df1)

df2


```

```python
plt.figure()

df1.hist()
```

```python
plt.figure()
df2.hist()
```

```python
data_copy['count'] = df2

data_copy
```

```python
#we can short the prevous by --- data_copy['count'] = np.log(data_copy['count'])
```

### Solve the problem of High Autocorrelation



```python
# Shift the count by 3 lags

dfNew1 = data_copy['count'].shift(+1).to_frame()  # 1 number of shifts

dfNew1.columns = ['Count_1']

dfNew2 = data_copy['count'].shift(+2).to_frame()  # 2 number of shifts

dfNew2.columns = ['Count_2']

dfNew3 = data_copy['count'].shift(+3).to_frame() # 3 number of shifts

dfNew3.columns = ['Count_3']


data_copy_lags = pd.concat([data_copy, dfNew1, dfNew2, dfNew3], axis= 1)
```

```python
data_copy_lags
```

```python
# count_1, count_2, count_3 === means three copy of count but in last times
```

```python
data_copy_lags = data_copy_lags.dropna()
```

```python
data_copy_lags
```

```python
# ----------------------------- Done ------------------------------------# 

#data_copy_lags['season'].value_counts()
```

### Now we can use the new data in predicting the count (the demand of the bikles)

```python
# we create Dummy variables  --- Convert all the columns to independent columns [Categorical Data Exactly] 
# such as convert the season to four columns
```

```python
# how to make this by using astype functon

data_copy_lags['season'] = data_copy_lags['season'].astype('category')
data_copy_lags['holiday'] = data_copy_lags['holiday'].astype('category')
data_copy_lags['weather'] = data_copy_lags['weather'].astype('category')
data_copy_lags['workingday'] = data_copy_lags['workingday'].astype('category')


data_copy_lags = pd.get_dummies(data_copy_lags, drop_first=True)
```

```python
data_copy_lags
```

### Now create train and Test

```python
# Count is time series type of data

Y = data_copy_lags['count'] # predicted coulumn
X = data_copy_lags.drop(['count'], axis=1)  # depended values (Other features)
```

### Fit  and Score the model

```python
# Linear Regresson
from sklearn.linear_model import LinearRegression
std_reg = LinearRegression()

# Training Process (Fitting) Ø¹Ù…Ù„ÙŠØ© Ø§Ù„ØªØ¯Ø±ÙŠØ¨
std_reg.fit(x_train,y_train)


# calculate the accuracy of Model
r_training = std_reg.score(x_train, y_train) # Accuracy of Training
r_testing = std_reg.score(x_test, y_test)  # Model Accuracy in Testing

r_training
```

```python
# Create Y predction

y_prediction = std_reg.predict(x_test)

y_prediction
```

```python
# Calculate Accuracy - another measure   --- Rout mean Square Error

from sklearn.metrics import mean_squared_error
rmsa_measure = math.sqrt(mean_squared_error(y_test, y_prediction))
rmsa_measure
```
